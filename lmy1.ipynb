```python
from qiskit import *
from qiskit.visualization import plot_histogram #plot_histogram:用来创建一个直方图
from numpy import pi
qc1=QuantumCircuit(8,8)#初始化变量 创建了一个有8个量子比特，8个经典比特输入的量子电路
```

```python
#白白
qc1.h(0)
qc1.h(1)
qc1.x(0)
qc1.x(1)
qc1.ccx(0,1,2)
qc1.ccx(0,1,3)
qc1.ccx(0,1,4)
qc1.ccx(0,1,6)
qc1.ccx(0,1,7)
qc1.x(0)
qc1.x(1)
qc1.x(0)
qc1.ccx(0,1,2)
qc1.ccx(0,1,5)
qc1.ccx(0,1,6)
qc1.ccx(0,1,7)
qc1.x(0)
qc1.x(1)
qc1.ccx(0,1,3)
qc1.ccx(0,1,4)
qc1.ccx(0,1,5)
qc1.ccx(0,1,7)
qc1.x(1)
qc1.ccx(0,1,4)
qc1.ccx(0,1,5)
qc1.ccx(0,1,6)
qc1.barrier()
for j in range(8):
    qc1.measure(j,j)
qc1.draw(output='mpl',scale=0.5,fold=100)  #使用MatPlotLib绘图
```

```python
qc1.draw(output='mpl').savefig('Fig6.1.pdf')
```

```python
backend_sim = Aer.get_backend('qasm_simulator')#指定模拟器
#job_sim = backend_sim.run(qc)
job_sim = execute(qc1,backend_sim, shots=100000)#执行量子线路
#execute 可以指定三个参数，分别是运行的量子线路名(circuit)、使用哪个模拟器(simulator)、执行多少次(shots，默认为 1024 次)
result_sim = job_sim.result()
counts = result_sim.get_counts()
plot_histogram(counts)#使用 plot_histogram 函数，以直方图的形式展示线路执行的结果
#可以通过 result.get_counts(circuit) 获取指定线路的运行结果
```

```python
plot_histogram(counts).savefig("Fig10.pdf",dpi=500,bbox_inches = 'tight')#解决图片不清晰，不完整的问题
```
